the above residual plot is not a good fit for a normal distribution. The residuals are not randomly scattered around the horizontal axis , but instead, they are clustered in the middle and taper off towards the ends. This is a sign of non-normality. The residuals are also not symmetric around the horizontal axis, which is another sign of non-normality.

Evalue the MEAN SQUARED ERROR AND R SQUARED SCORE S FOR THE MODEL. MSE measures the average of the squares of errors. that is the difference between actual value of y and the estimated value of y called yhat, while R squared measures the proportion of the variance for the dependent variable that’s explained by any independent variable , r squared is also called coeffient of determination is a measure to indicate how close the data is to the fittedd regression line, the value of r squared is the percentage of variation of the response variable or the target variable that is explained by the linear model.

```python
from sklearn.metrics import mean_squared_error 
```

```python
mse = mean_squared_error(df['card'], Yhat)    
mse
```
